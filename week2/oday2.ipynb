## Forcing dark mode

Gradio appears in light mode or dark mode depending on the settings of the browser and computer. There is a way to force gradio to appear in dark mode, but Gradio recommends against this as it should be a user preference (particularly for accessibility reasons). But if you wish to force dark mode for your screens, below is how to do it.

```python
# Define this variable and then pass js=force_dark_mode when creating the Interface

force_dark_mode = """
function refresh() {
    const url = new URL(window.location);
    if (url.searchParams.get('__theme') !== 'light') {
        url.searchParams.set('__theme', 'light');
        window.location.href = url.href;
    }
}
"""
gr.Interface(fn=shout, inputs="textbox", outputs="textbox", flagging_mode="never", js=force_dark_mode).launch()
```

```python
# Let's create a call that streams back results
# If you'd like a refresher on Generators (the "yield" keyword),
# Please take a look at the Intermediate Python guide in the guides folder

def stream_minimax(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = ollama.chat.completions.create(
        model=minimax,
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for minimax cloud", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_minimax,
    title="Minimax-Cloud", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Explain the Transformer architecture to a layperson",
        "Explain the Transformer architecture to an aspiring AI engineer",
        ], 
    flagging_mode="never"
    )
view.launch()
```

    The Transformer Architecture: A Comprehensive Guide
The Transformer, introduced in the landmark 2017 paper "Attention Is All You Need" by Vaswani et al., fundamentally reshaped deep learning. To understand why, let's start with the problem it solved.

The Problem With What Came Before
Before Transformers, sequence modeling (translation, text generation, etc.) was dominated by RNNs and LSTMs. These processed data sequentially, one token at a time. This created two major pain points:

Sequential bottleneck – Each step depended on the previous one, making training slow and hard to parallelize on GPUs.
Long-range dependencies – Information from earlier tokens had to travel through many steps, making it hard to "remember" distant context.
The Transformer took a radically different approach: process all tokens in parallel and let the model learn which tokens to attend to.

The Core Idea: Self-Attention
The heart of the Transformer is the self-attention mechanism. The question it answers is: "Given a word in a sentence, which other words in that sentence should it pay attention to in order to understand its meaning?"

The Query, Key, and Value Analogy
Imagine you're at a library looking for information about a topic (your query). Every book in the library has a label describing what it's about (its key). You compare your query against all the keys to find the most relevant books, then you read those books (their values) to extract information.

In mathematical terms, for each token, we compute three vectors by multiplying its embedding with three learned weight matrices:

Query (Q) – What am I looking for?
Key (K) – What do I contain?
Value (V) – What information do I provide?
The attention score between two tokens is computed as:

Attention(Q, K, V) = softmax(QKᵀ / √d_k) V

The dot product QKᵀ measures compatibility. We scale by √d_k to prevent the softmax from saturating. The softmax converts scores into a probability distribution, and we use those weights to take a weighted sum of the Value vectors.

What This Means Intuitively
For the sentence "The animal didn't cross the street because it was too tired", self-attention lets the model link "it" strongly to "animal", resolving the pronoun reference automatically based on context.

The Full Architecture
A Transformer has two main parts: an encoder and a decoder, each being a stack of identical layers.

Encoder Layer
Each encoder layer contains two sub-layers:

Multi-head self-attention – Instead of computing attention once, we run multiple attention "heads" in parallel, each learning to focus on different types of relationships (syntactic, semantic, positional, etc.). Their outputs are concatenated and projected.
Feed-forward network – A position-wise MLP applied independently to each token, typically with a hidden dimension 4× the model dimension and a ReLU/GELU activation.
Around each sub-layer, we have two crucial additions:

Residual connection – The input is added to the sub-layer's output. This helps gradients flow and makes deep networks trainable.
Layer normalization – Stabilizes training by normalizing activations.
Decoder Layer
The decoder has three sub-layers:

Masked multi-head self-attention – Same as the encoder, but with a mask that prevents the decoder from "looking ahead" at future tokens during training. This preserves the autoregressive property.
Cross-attention – The queries come from the decoder, but the keys and values come from the encoder's output. This is how the decoder "consults" the input sequence while generating output.
Feed-forward network – Same as in the encoder.
Positional Encoding
Because self-attention treats all tokens as a set with no inherent order, we need to inject position information. The original paper used fixed sinusoidal functions, though modern models often use rotary positional embeddings (RoPE) or ALiBi which generalize better to longer sequences.

Why Transformers Won
Transformers offer several decisive advantages:

Parallelism – All tokens are processed simultaneously, making training vastly more efficient on GPUs.
Long-range modeling – Self-attention has a constant path length between any two tokens (vs. O(n) for RNNs), making it easier to capture distant relationships.
Scalability – They scale predictably and well with more data and parameters, which led directly to the era of GPT, BERT, LLaMA, and friends.
Versatility – The same architecture works for text, images (ViT), audio (Whisper), protein folding (AlphaFold), code, and more.
From Vanilla Transformer to Modern LLMs
The original Transformer was an encoder-decoder model designed for translation. Today's large language models are typically:

Encoder-only (BERT-style) – Great for understanding tasks like classification and retrieval.
Decoder-only (GPT-style) – Great for generation, and the dominant architecture for modern LLMs.
Encoder-decoder (T5, BART) – Used for sequence-to-sequence tasks like translation and summarization.
Modern variants like LLaMA, Mistral, and Qwen use architectural refinements such as:

RMSNorm instead of LayerNorm
SwiGLU activations in the feed-forward layers
Grouped-query attention (GQA) for inference efficiency
Rotary positional embeddings (RoPE) with extended context windows
A Mental Model to Keep
Think of the Transformer as two main ingredients:

Self-attention – a learned, content-based routing mechanism that lets every token gather information from every other token.
Feed-forward layers – per-token "thinking" applied after information has been gathered.
Layer normalization and residual connections make the whole thing trainable at depth. Stack enough of these layers, train on enough data, and you get emergent capabilities that have changed the field.

    The Transformer is a type of artificial intelligence (AI) model that's been widely used for natural language processing tasks, like translation, text summarization, and question-answering. It was first introduced in 2017 by a team of researchers from Google.

Think of the Transformer as a powerful message decoder. Here's how it works:

The Problem with Traditional Models Traditional AI models use techniques called recurrent neural networks (RNNs) to process sequences like text or speech. RNNs work by looking at the previous messages to predict the next one. This is similar to trying to guess the next word in a sentence based on what came before it.

The Transformer's Innovative Approach The Transformer model, on the other hand, uses attention mechanisms to focus directly on specific parts of the input sequence rather than relying on prior knowledge. It's like tuning your radio to listen to only the station you care about – the one with the most relevant information.

Imagine a message being processed by a vast team of expert listeners, each focused on different aspects of the message. The Transformer's architecture allows these "listeners" (called attention heads) to communicate directly with each other and share knowledge from their specific areas of expertise. This results in more accurate predictions and much better results for natural language processing tasks.

How It Works

Input sequences are processed, line by line or word by word.
Each position in the sequence is compared against every other position to determine which parts of the message influence it most.
The comparison process produces a set of weights, similar to ratings of importance, that represent how relevant each part of the input sequence is to the current position.
These weights are applied to all positions in the sequence independently, allowing the model to weigh different aspects differently depending on where you're at in the message.
The final result is a more accurate prediction based on direct attention to relevant parts of the message.
What Makes It Special The Transformer's ability to focus selectively and respond effectively to specific parts of its input makes it extremely efficient, scalable, and easy to implement for complex language processing tasks.

## And now getting fancy

Remember to check the Intermediate Python Guide if you're unsure about generators and "yield"

```python
def stream_model(prompt, model):
    if model=="minimax":
        result = stream_minimax(prompt)
    elif model=="llama":
        result = stream_llama(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message for the LLM", lines=7)
model_selector = gr.Dropdown(["minimax", "llama"], label="Select model", value="GPT")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_model,
    title="LLMs", 
    inputs=[message_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Explain the Transformer architecture to a layperson", "minimax"],
            ["Explain the Transformer architecture to an aspiring AI engineer", "llama"]
        ], 
    flagging_mode="never"
    )
view.launch()
```

```python
from scraper import fetch_website_contents
```

```python

# Again this is typical Experimental mindset - I'm changing the global variable we used above:

system_message = """
You are an assistant that analyzes the contents of a company website landing page
and creates a short brochure about the company for prospective customers, investors and recruits.
Respond in markdown without code blocks.
"""
```

```python
def stream_brochure(company_name, url, model):
    yield ""
    prompt = f"Please generate a company brochure for {company_name}. Here is their landing page:\n"
    prompt += fetch_website_contents(url)
    if model=="minimax":
        result = stream_minimax(prompt)
    elif model=="llama":
        result = stream_llama(prompt)
    else:
        raise ValueError("Unknown model")
    yield from result
```

```python
name_input = gr.Textbox(label="Company name:")
url_input = gr.Textbox(label="Landing page URL including http:// or https://")
model_selector = gr.Dropdown(["minimax", "llama"], label="Select model", value="GPT")
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_brochure,
    title="Brochure Generator", 
    inputs=[name_input, url_input, model_selector], 
    outputs=[message_output], 
    examples=[
            ["Hugging Face", "https://huggingface.co", "minimax"],
            ["Edward Donner", "https://edwarddonner.com", "llama"]
        ], 
    flagging_mode="never"
    )
view.launch()
```
