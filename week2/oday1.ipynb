## Testing out the best models on the planet

````python
hard = """
On a bookshelf, two volumes of Pushkin stand side by side: the first and the second.
The pages of each volume together have a thickness of 2 cm, and each cover is 2 mm thick.
A worm gnawed (perpendicular to the pages) from the first page of the first volume to the last page of the second volume.
What distance did it gnaw through?
"""
hard_puzzle = [
    {"role": "user", "content": hard}
]
````

````python
#response = ollama.chat.completions.create(model="minimax-m2:cloud", messages=hard_puzzle, reasoning_effort="minimal")
response = ollama.chat.completions.create(model="minimax-m3:cloud", messages=hard_puzzle, reasoning_effort="medium")

display(Markdown(response.choices[0].message.content))
````

minimax-m3:cloud 1m 2.2s : Looking at this problem, I need to figure out the actual path the worm takes through the books.

## Setting Up the Geometry

When two volumes stand side-by-side on a shelf in proper order (Volume 1 on the left, Volume 2 on the right), their spines face outward and their **front covers face each other**.

**Volume 1 (spine on left):**
- Back cover ← Last page ← ... → First page → Front cover
- The **first page sits on the right edge**, just inside the front cover

**Volume 2 (spine on right):**
- Front cover → First page → ... → Last page → Back cover
- The **last page sits on the right edge**, just inside the back cover

## Tracing the Worm's Path

The worm travels from the first page of Volume 1 to the last page of Volume 2:

| Section | Distance |
|---------|----------|
| Front cover of Volume 1 | 2 mm |
| Front cover of Volume 2 | 2 mm |
| All pages of Volume 2 | 20 mm |
| **Total** | **24 mm** |

## The Key Insight

The worm **does not** gnaw through:
- The pages of Volume 1 (it starts at the first page, which is on the right side)
- The back cover of Volume 1 (on the opposite side from where the worm enters/exits)
- The back cover of Volume 2 (it ends at the last page, which is just inside this cover)

## Answer

The worm gnawed through **2.4 cm** (24 mm).

````python
response = ollama.chat.completions.create(model="llama3.2:3b", messages=hard_puzzle)
display(Markdown(response.choices[0].message.content))
````

llama3.2:3b : To find the total distance traveled by the worm, let's break down the problem step by step.

1. The total thickness of the two volumes is 2 cm = 20 mm (since there are 10 mm in a cm).
2. Since each cover is 2 mm thick, the total thickness of both covers combined is 2 x 3 = 6 mm (because each volume has one cover on each side). 
   So, after removing this section, the pages will be left.
3. The height that the worm moves through the bookshelf is 20 - 6= 14mm

````python
response = ollama.chat.completions.create(model="phi3", messages=hard_puzzle)
display(Markdown(response.choices[0].message.content))
````

phi3 2m 14.2s : To solve this problem we need to calculate two things; The total number of cover-back pairs and their thickness plus that for just a single set of pages (which will be common between both volumes). But since all covers are the same in each volume, what affects one is equal to itself times its quantity.

Firstly, we find out how many sets of backings there would have been if it had chewed through them on every page without stopping: One pair per two pages for a set within the volumes + shared back cover between both books. There'd be (total number of pages - 1) /2 pairs = ((2 cm/cover-back thickness difference *960 mm^3+8 mm)/thickness)^(2/3)-1)/2 which simplifies to approximately \(54\), or half of the total minus one due to shared back cover.

Now that we have figured out it would take about 54 sets for a complete journey through both volumes covering all pages and common covers, but considering there's no need to chew through them twice since they overlap (one set with only two books), our distance is found at \( \frac{Cover back-thickness + page thickness}{2} * 54 \) = ((1.38 mm - 0.2mm) /2)*54 which gives approximately a worm's journey of around 19 cm, as they share both ends and the cover on either end cut through once only (totaling at two covers).

## A spicy challenge to test the competitive spirit

````python
dilemma_prompt = """
You and a partner are contestants on a game show. You're each taken to separate rooms and given a choice:
Cooperate: Choose "Share" — if both of you choose this, you each win $1,000.
Defect: Choose "Steal" — if one steals and the other shares, the stealer gets $2,000 and the sharer gets nothing.
If both steal, you both get nothing.
Do you choose to Steal or Share? Pick one.
"""

dilemma = [
    {"role": "user", "content": dilemma_prompt},
]
````

````python
response = ollama.chat.completions.create(model=minimax, messages=dilemma)
display(Markdown(response.choices[0].message.content))
````

Minimax : I'll Share.

Here's my honest reasoning, even though the game theorist in me knows it's not the "optimal" play:

The math says steal. Defecting is the dominant strategy in a one-shot Prisoner's Dilemma — no matter what my partner does, stealing leaves me at least as well off. If they share, I get $2,000 instead of $1,000. If they steal, I get $0 either way.

But I don't think the game theory fully applies here. A few reasons:

I can't win alone. The best possible outcome for me as an individual is $2,000, but the best guaranteed good outcome is $1,000 — and that only happens through mutual cooperation.

The question feels like it's testing something deeper. The whole framing — separate rooms, a partner I never see, a single choice — is designed to surface what kind of actor I am. Picking "Steal" because the math allows it is the move of someone who treats every interaction as a one-off. That's a bleak way to move through the world.

Trust is cheap to extend and expensive to suspect. The cost of being wrong (sharing while they steal) is $0 instead of $1,000. The cost of being right to distrust is $0 instead of $1,000 — same outcome. But the upside of mutual trust is real and only accessible if at least one of us is willing to take the risk.

So: Share. Not because it's strategically smart, but because the version of me I want to be is the one who'd rather lose $1,000 trusting a stranger than win $2,000 assuming the worst of them.

What would you choose?

## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
# Let's make a conversation between GPT-4.1-mini and Claude-haiku-4.5
# We're using cheap versions of models so the costs will be minimal

gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5"

gpt_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

claude_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

gpt_messages = ["Hi there"]
claude_messages = ["Hi"]
````

````python
def call_gpt():
    messages = [{"role": "system", "content": gpt_system}]
    for gpt, claude in zip(gpt_messages, claude_messages):
        messages.append({"role": "assistant", "content": gpt})
        messages.append({"role": "user", "content": claude})
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_gpt()
````

````python
def call_claude():
    messages = [{"role": "system", "content": claude_system}]
    for gpt, claude_message in zip(gpt_messages, claude_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": claude_message})
    messages.append({"role": "user", "content": gpt_messages[-1]})
    response = anthropic.chat.completions.create(model=claude_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_claude()
````

````python
call_gpt()
````

````python
gpt_messages = ["Hi there"]
claude_messages = ["Hi"]

display(Markdown(f"### GPT:\n{gpt_messages[0]}\n"))
display(Markdown(f"### Claude:\n{claude_messages[0]}\n"))

for i in range(5):
    gpt_next = call_gpt()
    display(Markdown(f"### GPT:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)
    
    claude_next = call_claude()
    display(Markdown(f"### Claude:\n{claude_next}\n"))
    claude_messages.append(claude_next)
````
